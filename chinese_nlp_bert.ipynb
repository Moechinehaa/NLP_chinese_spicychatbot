```python
import torch
from datasets import load_dataset

dataset = load_dataset('text', data_files={'train': 'train.tsv', 'test': 'test.tsv'})
```

```python
dataset
```

```python
def remove_the_title(example,idx):
  return idx>0
dataset_value = dataset.filter(remove_the_title, with_indices=True, load_from_cache_file=False)
dataset_value
```

```python
def tokenize_function(examples):
    labels = []
    texts = []
    for example in examples['text']:
      split = example.split('\t', maxsplit=1)
      #print(split)
      labels.append(int(split[0]))
      texts.append(split[1])
    tokenized = tocken(texts, padding='max_length', truncation=True, max_length=32)
    tokenized['labels'] = labels
    return tokenized

tokenized_datasets = dataset_value.map(tokenize_function, batched=True)
train_dataset = tokenized_datasets["train"].shuffle(seed=42)
eval_dataset = tokenized_datasets["test"].shuffle(seed=42)
```

```python
from transformers import (AutoModelForSequenceClassification, AutoTokenizer,
                          Trainer, TrainingArguments)
```

```python
model = AutoModelForSequenceClassification.from_pretrained("bert-base-chinese", num_labels=8, cache_dir='pretrained')
training_args = TrainingArguments('output', per_device_train_batch_size=50, num_train_epochs=5)
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset
)
trainer.train()
```

```python
trainer.save_model("chinese_emotion_model")
```

```python
result = trainer.evaluate()
print(result)
```

```python
from datasets import load_metric
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import accuracy_score
import numpy as np

predictions = trainer.predict(eval_dataset)

def compute_metrics(pred):
  labels = pred.label_ids
  preds = pred.predictions.argmax(-1)
  #precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='binary')
  acc = accuracy_score(labels, preds)
  print(acc)
```
